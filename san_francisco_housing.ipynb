```python
# Using the read_csv function and Path module, create a DataFrame 
# by importing the sfo_neighborhoods_census_data.csv file from the Resources folder
sfo_data_df = pd.read_csv(
    Path('Starter_Code/sfo_neighborhoods_census_data.csv')
)



# Review the first five rows of the DataFrame
sfo_data_df
```

```python
# Load neighborhoods coordinates data
neighborhood_locations_df = pd.read_csv(
    Path('Starter_Code/neighborhoods_coordinates.csv')
)
neighborhood_locations_df
```

```python
#Combined DataFrame for 2 data sets e.g. sfo_data & neighborhood_locations_
all_neighborhood_info_df = pd.concat([sfo_data_df, neighborhood_locations_df], axis="columns", join="inner")
# view the resulting DataFrame
all_neighborhood_info_df


# YOUR CODE HERE
```

```python
# Calculate the mean values for each neighborhood
neighborhood_info_by_year = all_neighborhood_info_df.groupby('neighborhood').mean()
neighborhood_info_by_year
```

```python
# Calculate the mean values for each neighborhood
df_neghborhood = neighborhood_info_by_year[['sale_price_sqr_foot','gross_rent','Lon','Lat']]
df_neghborhood
```

```python
# Plot the data using hvPlot and Geoviews
map_plot = df_neghborhood.hvplot.points(
    'Lon', 
    'Lat', 
    geo=True, 
    size = 'sale_price_sqr_foot',
    #scale = .02,
    color='gross_rent',
    #alpha=0.8,
    tiles='OSM',
    frame_width = 700,
    frame_height = 500
    ).opts(yformatter = '%.0f')

map_plot
```
